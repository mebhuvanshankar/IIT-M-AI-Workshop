```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(0,10,101)
x=[1,2,3,4,5]
y1=np.sin(x)
plt.plot(x,y1)
plt.title('subplot')
plt.ylabel('y1 axis')
plt.show()
```

```python
y2=np.cos(x)
plt.plot(x,y2)
plt.title('subplot')
plt.ylabel('y2 axis')
```

```python
y3=np.tan(x)
plt.plot(x,y3)
plt.title('subplot')
plt.ylabel('y3 axis')
```

```python
y4=np.exp(x)
plt.plot(x,y4)
plt.title('subplot')
plt.ylabel('y4 axis')
```
